```python
import pandas as pd
import numpy as np
```

```python
olivos = pd.read_csv('./olivos.csv')
```

```python
del olivos["Unnamed: 0"]
del olivos["number"]
olivos.head()
```

```python
data = []
for i, row in olivos.iterrows():
    if (row["fin"] == "JEFATURA"):
        nombre = row["nombre"]
        documento = row["doc"]
        hora = row["hora"]
        dia = row["dia"]
        (year, month, day) = dia.split("-")
        if (int(year) != 2020 and int(year) != 2021):
            (prev_year, prev_month, prev_day) = olivos["dia"][i - 1].split("-")
            dia = f"{prev_year}-{month}-{day}"
        data.append({ "nombre": nombre, "documento": documento, "dia": dia, "hora": hora })
```

```python
df = pd.DataFrame(data)
```

```python
fechas = []
for dia_hora, ids in df.groupby(["dia", "hora"]).groups.items():
    (dia, hora) = dia_hora
    fechas.append(f"{dia}/{hora}")
```

```python
documentos = pd.unique(df["documento"])
```

```python
obj = { "fechas": fechas }
```

```python
len_docs = len(documentos)
len_fechas = len(fechas)
i = 0
for doc in documentos:
    visito = []
    for f in fechas:
        (dia, hora) = f.split("/")
        fue = int(((df["documento"] == doc) & (df["dia"] == dia) & (df["hora"] == float(hora))).any())
        visito.append(fue)
    obj[doc] = visito
    i += 1
    print(i / len_docs * 100)
```

```python
def getName(doc):
    r = olivos.loc[olivos["doc"] == doc]["nombre"].head(1)
    i = r.keys()[0]
    return r[i]
```

```python
dfm = pd.DataFrame(obj).set_index("fechas")
ds_asint = dfm.astype(int)
coocc = ds_asint.T.dot(ds_asint)
np.fill_diagonal(coocc.values, 0)
coocc
```

```python
doc_to_name = []
for k in coocc.keys():
    if (str(k) != "nan"):
        doc_to_name.append(getName(k))
    else:
        print(k)
        doc_to_name.append(str(k))
```

```python
coocc.columns = doc_to_name
coocc.index = doc_to_name
coocc
```

```python
coocc.to_csv('./matrix.csv')
```
